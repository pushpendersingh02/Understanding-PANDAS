# JOIN

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'A':[1,2,3,4,5,6],'B':[9,8,7,6,5,4],'C':[90,80,70,60,50,40]},index=['Aman','Axar','Bobbie','Chaman','Rollex','Simon'])
df1
```

```python
df2=pd.DataFrame({'D':[10,20,30,40,50,60],'E':[19,18,17,16,15,14],'F':[91,82,73,64,55,46]},index=['Aman','Axar','Bobbie','Chaman','Rollex','Simon'])
df2
```

```python
df1.join(df2)
```

```python
df3=pd.DataFrame({'D':[10,20,30,40],'E':[19,18,17,16],'F':[91,82,73,64]},index=['Axar','Akash','Bobbie','Chaman'])
df3
```

```python
df1.join(df3)
```

```python
df3.join(df1)
```

#### How Parameter

```python
df3.join(df1, how='right')
```

```python
df3.join(df1, how='left')
```

```python
df3.join(df1, how='inner')
```

```python
df3.join(df1, how='outer')
```

#### Joining DataFrames if they have same column names

```python
df3.join(df2)
```

```python
df3.join(df2, how='outer',lsuffix='_Set1')
```

```python
df3.join(df2, how='outer',lsuffix='_Set1', rsuffix='_Set2')
```

```python
df3.join(df2, how='inner',lsuffix='_Set1', rsuffix='_Set2')
```
