# Merge

```python
import pandas as pd
```

```python
var1={'A':[1,20,3,4,5],'B':[11,12,13,14,15]}
var2={'A':[1,2,3,4,5],'D':[31,32,33,34,35]}
df1=pd.DataFrame(var1,index=['Delhi','Lucknow','Gujrat','Chennai','Mumbai'])
df1
```

```python
df2=pd.DataFrame(var2,index=['Delhi','Lucknow','Gujrat','Chennai','Mumbai'])
df2
```

```python
pd.merge(df1,df2)
```

```python
 pd.merge(df1,df2,on ='A')
```

```python
pd.merge(df1,df2,how='inner')
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,how='outer',indicator=True)
```

```python
var3={'A':[1,2,3,4,5],'B':[11,12,13,14,15]}
var4={'A':[1,2,3,4,5],'B':[31,32,33,34,35]}
df3=pd.DataFrame(var3,index=['Delhi','Lucknow','Gujrat','Chennai','Mumbai'])
df3
```

```python
df4=pd.DataFrame(var4,index=['Delhi','Lucknow','Gujrat','Chennai','Mumbai'])
df4
```

```python
pd.merge(df3,df4)
# it's showing empty because column names are same in both DataFrames, we have to use left and right index here
```

```python
pd.merge(df3,df4,left_index=True,right_index=True)
```

```python
pd.merge(df3,df4,left_index=True,right_index=True,suffixes=[' Name',' Id'])
```

# Concat

```python
sr1=pd.Series([1,2,3,4,5,6])
sr1
```

```python
sr2=pd.Series([11,21,31,41,51,61])
sr2
```

```python
pd.concat([sr1,sr2])
```

#### Diff between MERGE and CONCAT in case of DataFrame

```python
pd.concat([df1,df2])
# It will just add 2 DataFrames irrespective of checking any common row/column
```

##### ignore_index Parameter

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.merge(df1,df2)
# It will merge the 2 DataFrames on common column/row in the DataFrames
```

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=1)
```

```python
v1={'A':[1,2,3,4,5],'B':[11,12,13,14,15]}
v2={'C':[1,2,3],'D':[31,32,33]}
df5=pd.DataFrame(v1,index=['Delhi','Lucknow','Gujrat','Chennai','Mumbai'])
df5
```

```python
df6=pd.DataFrame(v2,index=['Delhi','Chennai','Mumbai'])
df6
```

```python
pd.concat([df5,df6],axis=1)
```

#### Join Parameter

```python
pd.concat([df5,df6],axis=1, join='inner')
# it shows intersection of 2 DataFrames
```

```python
pd.concat([df5,df6],axis=1,join='outer')
#it shows union of 2 DataFrames
```

```python
pd.concat([df1,df2,df3,df4,df5,df6],axis=1)
```

#### Keys Parameter

```python
pd.concat([df3,df4],axis=1,keys=['D1','D2'])
```

```python
pd.concat([df3,df4],axis=0,keys=['D1','D2'])
```

#### Concatinating 1D and 2D DataFrames

```python
d1=pd.DataFrame({'A':[1,2,3,4,5,6]})
# 1D DataFrame
d1
```

```python
d2=pd.DataFrame({'B':[10,20,30,40,50,60],'C':[9,8,7,6,5,4]})
# 2D DataFrame
d2
```

#### ignore_index Parameter

```python
pd.concat([d1,d2])
```

```python
pd.concat([d1,d2],axis=1)
```
