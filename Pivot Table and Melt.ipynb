# MELT()

```python
# It will reshape your table/DataFrame vertically
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Days":[1,2,3,4,5,6,7],"English":[23,54,11,64,66,53,23],"Maths":[29,11,26,37,45,40,59]})
df
```

```python
pd.melt(df)
```

```python
pd.melt(df,id_vars='Days')
```

```python
pd.melt(df,id_vars=['Days'], var_name='Subject')
```

```python
pd.melt(df,id_vars=['Days',"Maths"], var_name='Subject')
```

```python
pd.melt(df,id_vars=['Days'], var_name='Subject',value_name='Marks')
```

# PIVOT()

```python
pv = pd.DataFrame({"Days":[1,2,3,4,5,6,7],'Names':['Rahul','Rahul','Arvind','Modi','Rahul','Arvind','Modi'],"English":[23,54,11,64,66,53,23],"Maths":[29,11,26,37,45,40,59]})
pv
```

```python
pv.pivot(index='Days', columns='Names')
```

```python
pv.pivot(index='Days', columns='Names',values='English')
```

# PIVOT TABLE

```python
# We can use aggregate functions in pivot_table
```

```python
pv1 = pd.DataFrame({"Days":[1,1,1,1,2,2,2],'Names':['Rahul','Rahul','Arvind','Modi','Rahul','Arvind','Modi'],"English":[23,54,11,64,66,53,23],"Maths":[29,11,26,37,45,40,59]})
pv1
```

```python
pv1.pivot_table(index='Names', columns='Days',
               aggfunc='mean')
```

```python
pv1.pivot_table(index='Names', columns='Days',
               aggfunc='max')
```

```python
pv1.pivot_table(index='Names', columns='Days',
               aggfunc='min')
```

```python
pv1.pivot_table(index='Names', columns='Days',
               aggfunc='sum')
```
