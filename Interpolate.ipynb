# Interpolate

```python
import pandas as pd
```

```python
csv=pd.read_csv('Trial1.csv')
csv
```

```python
csv.interpolate()
# This function only works on Numbers and not on Alphabets, so numbers will be populated but not alphabets.
```

#### Interpolate Methods: {'linear', 'time', 'index', 'values', 'nearest', 'zero', 'slinear', 'quadratic', 'cubic', 'barycentric', 'krogh', 'polynomial', 'spline', 'piecewise_polynomial', 'from_derivatives', 'pchip', 'akima'}

```python
csv.interpolate(method='linear')
```

#### Axis

```python
# axis=1, it will fill data row by row. The complete table must have same datatype to use this.
# axis=0, it will fill data column by column

csv.interpolate(axis=0)
```
