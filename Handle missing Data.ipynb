```python
csv1=pd.read_csv('Trial2.csv')
csv1
```

```python
csv1.dropna(how='all')
#only removes the row which is completely NaN
```

```python
csv1.dropna(how='any')
#removes all the rows which contains atleast 1 value as NaN
```

```python
csv1.dropna(how='all',axis=1)
```

```python
csv1.dropna(how='any',axis=1)
```

```python
csv1.dropna(subset=['Marks_Sub3'])
#removes the rows where null is present in Marks_Sub3 column.
```

```python
csv1.dropna(inplace=True)
csv1
```

```python
csv1=pd.read_csv("Trial2.csv")
csv1
```

```python
csv1.dropna(thresh = 1)
```

# fillna

```python
csv1.fillna(method='ffill')
```

```python
csv1.fillna(method='bfill')
```

```python
csv1.fillna(method='ffill',limit=1)
```

```python
csv1.fillna(method='ffill',limit=2)
```

```python
csv1.fillna(method='ffill', axis=1,limit=1)
```
